```python
import numpy as np

def linear_trans(x, w, b):
    '''
    Linear transfromation.
    
    Args:
        x:    Observed data points.
        w:    Weight matrix.
        b:    bias vector.
    
    Returns:
        y.
    '''
    y = np.dot(x, w) + b
    return y

x = np.random.normal(size=(20, 5))
w = np.random.normal(size=(5, 10))
b = np.random.normal(size=(10))

y = linear_trans(x, w, b)
y
```

```python
import numpy as np

def compute_mse(pred_arr, real_arr, ax=0):
    '''
    Mean Squared Error.
    
    Args:
        pred_arr:    `np.ndarray` of predicted data.
        real_arr:    `np.ndarray` of real data.
    
    Returns:
        MSE.
    '''
    return (np.square(pred_arr - real_arr)).mean(axis=ax)
```

```python
pred_arr = np.random.normal(loc=0.1, scale=0.01, size=(100, 1))
real_arr = np.random.normal(loc=0.6, scale=0.01, size=(100, 1))

compute_mse(pred_arr, real_arr)
```
