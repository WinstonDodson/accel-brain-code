```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
def sigmoid(x):
    '''
    Sigmoid function.
    
    Args:
        x:    Observed data points.
    
    Returns:
        y
    '''
    return 1 / (1 + np.exp(-x))
```

```python
x1 = np.arange(-10.0, 10.0, 0.1)
y1 = sigmoid(x1)

plt.plot(x1, y1)
plt.xlabel("Observed data points.")
plt.ylabel("Activation.")
plt.title("The sigmoid function.")
plt.show()
```

```python
def derivative_sigmoid(x):
    '''
    derivative.
    
    Args:
        x:    observed data points.
    
    Returns:
        y.
    '''
    return (1 - sigmoid(x)) * sigmoid(x)
```

```python
y1 = sigmoid(x1)
d1 = derivative_sigmoid(x1)

plt.plot(x1, y1, label="Activation.")
plt.plot(x1, d1, label="derivative.")
plt.xlabel("Observed data points.")
plt.ylabel("The results.")
plt.legend()
plt.title("The sigmoid function.")
plt.show()
```

```python
def tanh(x):
    '''
    tanh function.
    
    Args:
        x:    Observed data points.
    
    Returns:
        y
    '''
    return np.tanh(x)
```

```python
x2 = np.arange(-10.0, 10.0, 0.1)
y2 = tanh(x2)

plt.plot(x2, y2)
plt.xlabel("Observed data points.")
plt.ylabel("Activation.")
plt.title("The tanh function.")
plt.show()
```

```python
def derivative_tanh(x):
    '''
    derivative.
    
    Args:
        x:    observed data points.
    
    Returns:
        y.
    '''
    return 1 - (tanh(x) ** 2)
```

```python
x2 = np.arange(-10.0, 10.0, 0.1)
y2 = tanh(x2)
d2 = derivative_tanh(x2)

plt.plot(x2, y2, label="Activation.")
plt.plot(x2, d2, label="Derivative.")
plt.xlabel("Observed data points.")
plt.ylabel("The results.")
plt.title("The tanh function.")
plt.legend()
plt.show()
```

```python
def relu(x):
    '''
    ReLu function.
    
    Args:
        x:    Observed data points.
    
    Returns:
        y
    '''
    return np.maximum(x, 0)
```

```python
x3 = np.arange(-10.0, 10.0, 0.1)
y3 = relu(x3)

plt.plot(x3, y3)
plt.xlabel("Observed data points.")
plt.ylabel("Activation.")
plt.title("The ReLu function.")
plt.show()
```

```python
import numpy as np

def linear_trans(x, w, b):
    '''
    Linear transfromation.
    
    Args:
        x:    Observed data points.
        w:    Weight matrix.
        b:    bias vector.
    
    Returns:
        y.
    '''
    y = np.dot(x, w) + b
    return y


# Init(Hidden layer).
w1 = np.random.normal(size=(5, 10))
b1 = np.random.normal(size=(10))
# Init(Output layer).
w2 = np.random.normal(size=(10, 5))
b2 = np.random.normal(size=(5))

# Observed data points.
x1 = np.random.normal(size=(20, 5))

# Forward in hidden layer.
x2 = linear_trans(x1, w1, b1)
# Activation in hidden layer.
x2 = sigmoid(x2)

# Forward in output layer.
x3 = linear_trans(x2, w2, b2)

# Activation in output layer.
y = sigmoid(x3)
y
```
